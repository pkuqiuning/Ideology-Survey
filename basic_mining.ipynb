```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
data = pd.read_csv('2014data_clean_geoip.csv')
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
y=data.latitude
x=data.longitude
```

```python
def draw_hexbin(x, y, fig_filename='dist.png', gridsize=100, crop=False, lw=1, alpha=1):
    # draw hexbin of x, y on top of a map
    # modified code from http://sensitivecities.com/so-youd-like-to-make-a-map-using-python-EN.html#.VR7QreQx7XY
    plt.gcf().set_size_inches(16,10)
    if crop:
        m = Basemap(llcrnrlon=crop[0], llcrnrlat=crop[1], urcrnrlon=crop[2], urcrnrlat=crop[3])
    else:
        m = Basemap()
    m.drawlsmask(land_color='white', ocean_color='lightgray')
    m.drawcountries()
    plt.hexbin(x, y,
           gridsize=gridsize,
           bins='log',
           mincnt=1,
           #edgecolor='none',
           alpha=alpha,
           lw=lw,
           cmap=plt.get_cmap('cool'))
    plt.colorbar(shrink=0.6).set_label('log10(N)')
    plt.savefig(fig_filename,dpi=100)

draw_hexbin(x, y, 'dist_world.png', gridsize=300, lw=0.5)
```

```python
draw_hexbin(x[data.country=='China'],
            y[data.country=='China'],
            'dist_china.png',
            gridsize=100,
            lw=4.,
            alpha=0.6,
            crop=(70, 15, 140, 60))
```

```python
draw_hexbin(x[data.country=='United States'],
            y[data.country=='United States'],
            'dist_usa.png',
            gridsize=300,
            lw=4.,
            alpha=0.6,
            crop=(-140, 23, -60, 55))
```
